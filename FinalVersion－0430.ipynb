Scalable K-Means++
----

### Background

Clustering is a central problem in data management and has a rich and illustrious history with literally hundreds of different algorithms published on the subject. Even so, a single method — k-means — remains the most popular clustering method; in fact, it was identified as one of the top 10 algorithms in data mining. The advantage of k-means is its simplicity: starting with a set of randomly chosen initial centers, one repeatedly assigns each input point to its nearest center, and then recomputes the centers given the point assignment. This local search, called Lloyd’s iteration, continues until the solution does not change between two consecutive rounds.

K-means++ clustering is identical to the K-means algorithm, except for the selection of initial conditions.

In this paper the autors obtained a parallel version of the k-means++ initialization algorithm and empirically demonstrated its practical effectiveness. The main idea was that instead of sampling a single point in each pass of the k-means++ algorithm, they sampled O(k) points in each round and repeat the process for approximately O(log n) rounds. At the end of the algorithm they were left with O(k log n) points that formed a solution that was within a constant factor away from the optimum. They then reclustered these O(k log n) points into k initial centers for the Lloyd’s iteration. This initialization algorithm, which they called k-means||, was quite simple and lended itself to easy parallel implementations. However, the analysis of the algorithm turned out to be highly non-trivial, requiring new insights, and was quite different from the analysis of k-means++. They then evaluated the performance of this algorithm on real-world datasets. Their key observations in the experiments were: 
- O(log n) iterations was not necessary and after as little as five rounds, the solution of k-means|| was consistently as good or better than that found by any other method.
- The parallel implementation of k-means|| was much faster than existing parallel algorithms for k-means. 
- The number of iterations until Lloyd’s algorithm converges was smallest when using k-means|| as the seed.

### The algorithm

$k$-means|| ($k$, $\mathscr{l}$):

- $\mathcal{C} \leftarrow$ sample a point uniformly at random from $X$, where $X$ is the dataset.
- $\psi \leftarrow \phi x (\mathcal{C})$:
    - $\phi x (\mathcal{C})$ is the sum of all smallest euclidean distances from all points from $X$ to all points from $\mathcal{C}$. In other words, for each point in $X$ find the distance to the closest point in $\mathcal{C}$, and in the end compute the sum of all those minimal distances, one for each point in $X$.
    - So rewrite $\psi$ as $\psi = \sum_{i = 1}^{n}d_{c}^{2}(X_{i})$, where $d_{c}^{2}(X_{i})$ is the distance from $x_{i}$ to the closest point in $\mathcal{C}$.
- **for** $\mathcal{O}(\log \psi)$ times **do**:
    - $\mathcal{C}^{'} \leftarrow$ sample each point $x \in X$ independently with probability $p_{x} = \frac{\mathscr{l} d^{2}(x, \mathcal{C})}{\phi x (\mathcal{C})}$. 
    - $\mathcal{C} \leftarrow \mathcal{C} \cup \mathcal{C}^{'}$
- **end for**
    - In this for loop, $\phi x (\mathcal{C})$ is defined in previous step, and the rule is that, iterate in $X$ to find all $x_{i}$, for each $x_{i}$ compute $p_{x_{i}}$, generate an uniform number in [0,1] and if it's smaller than $p_{x_{i}}$, select it to form $\mathcal{C}^{'}$, and then after all draws are done, include selected points from $\mathcal{C}^{'}$ into $\mathcal{C}$. Note that the drawings are independent.
- For $c \in \mathcal{C}$, set $w_{c}$ to be the number of points in $X$ closer to $c$ than any other points in $\mathcal{C}$.
    - Create a vector $w$ of size equals to the number of elements in $\mathcal{C}$, and initialize all its values with 0. Iterate in $X$ (elements not selected in as centroids), and for each $x_{i} \in X$, find the index $j$ of the closest centroid (element from $\mathcal{C}$) and increment $w[j]$ with 1. In the end the vector $w$ can be computed properly.
- Recluster the weighted points in $\mathcal{C}$ into $k$ clusters。
    - Follow $k$-means++ algorithm to select only $k$ points as starting centroids. Thus, k for loops will be executed and at each loop select a single element drawn randomly with probability for each element being $p_{i} = \frac{w_{i}}{\sum_{j = 1}^{m}w_{j}}$. At each step select one element, and remove it from candidates, also removing its corresponding weight.


### Implementation

```python
import numpy as np
import sys

# define a function to return initial centroids for basic k means algorithm
def k_initial_centroids(data, k):
    # get the initial set of centroids
    # get k random numbers between 0 and the number of rows in the data set
    centroid_id = np.random.choice(range(data.shape[0]), k, replace = False)
    # get the corresponding data points
    return data[centroid_id, :]

# define a function to return initial centroids for k means ++ algorithm
def kpp_initial_centroids(data, k):
    # pick the initial centroid randomly
    centroids = data[np.random.choice(range(data.shape[0]), 1), :]
    data_new = data[:, np.newaxis, :]
    # run k - 1 passes through the data set to select the initial centroids
    while centroids.shape[0] < k:
        euclidean_dist = (data_new - centroids) ** 2
        distance = np.sum(euclidean_dist, axis = 2)
        min_id = np.zeros(distance.shape)
        min_id[range(distance.shape[0]), np.argmin(distance, axis=1)] = 1
        # calculate phi
        phi = np.sum(distance[min_id == True])
        # calculate the probability distribution
        prob = np.min(distance, axis=1) / phi
        # select the next centroid using the probability distribution calculated
        centroids = np.vstack([centroids, data[np.random.choice(range(data.shape[0]), 1, p = prob), :]])
    return centroids

# define a function to return initial centroids and data for scalable k means ++ algorithm
def scalable_kpp_initial_centroids(data, k, l):
    # Pick the initial centroid randomly
    centroids = data[np.random.choice(range(data.shape[0]),1), :]
    data_new = data[:, np.newaxis, :]    
    euclidean_dist = (data_new - centroids) ** 2
    distance = np.sum(euclidean_dist, axis=2)
    min_id = np.zeros(distance.shape)
    min_id[range(distance.shape[0]), np.argmin(distance, axis=1)] = 1
    phi_val = np.sum(distance[min_id == True])

    for i in range(int(round(np.log(phi_val)))):
        euclidean_dist_1 = (data_new - centroids) ** 2
        distance_1 = np.sum(euclidean_dist_1, axis=2)
        min_id_1 = np.zeros(distance_1.shape)
        min_id_1[range(distance_1.shape[0]), np.argmin(distance_1, axis=1)] = 1
        min_dist = distance_1[min_id_1 == True]
        phi_val = np.sum(min_dist)
        for j, p in enumerate(data):
            prob = l * min_dist[j] / phi_val
            u = np.random.uniform(0,1)
            if prob >= u:
                centroids = np.vstack([centroids, p])
    
    # Now we have the set of initial centroids that is greater than k
    # we should reduce this to k using scalable K-Means++
    euclidean_dist_2 = (data_new - centroids) ** 2
    distance_2 = np.sum(euclidean_dist_2, axis=2)
    min_id_2 = np.zeros(distance_2.shape)
    min_id_2[range(distance_2.shape[0]), np.argmin(distance_2, axis=1)] = 1
    weights = np.array([np.count_nonzero(min_id_2[:, j]) for j in range(centroids.shape[0])],dtype='float64').reshape(-1,1)
    
    #Recluster the weighted points into k clusters using kmeans++
    new_centroids = centroids[np.random.choice(range(centroids.shape[0]),1), :]
    old_centroids = centroids
    index = np.where(old_centroids == new_centroids)[0]
    old_centroids = np.delete(old_centroids, index[0], axis = 0)
    weights = np.delete(weights, index[0])
    while new_centroids.shape[0] < k:
        #Calculate the probability distribution based on weights
        prob = weights/np.sum(weights)
        #Select the next centroid using the prob calculated
        c = old_centroids[np.random.choice(range(old_centroids.shape[0]),1, p = prob.ravel()), :]
        index = np.where(old_centroids == c)[0]
        new_centroids = np.vstack([new_centroids, c])
        #Remove the selected center and its corresponding weight
        old_centroids = np.delete(old_centroids,index[0],axis=0)
        weights = np.delete(weights,index[0])
        
    return new_centroids, centroids


# define a function to compare the old centroids with new centroids to see if they converge
def compare_centroids(old_centroids, new_centroids, precision = -1):
    if precision == -1:
        return np.array_equal(old_centroids, new_centroids)
    else:
        diff = np.sum((new_centroids - old_centroids)**2, axis = 1)
        if np.max(diff) <= precision:
            return True
        else:
            return False
   
        
# define a function to implement the k means algorithm    
def kmeans(centroids, data, k):
    # for basic k means algorithm, centroids = k_initial_centroids(data, k)
    # for k means pp algorithm, centroids = kpp_initial_centroids(data, k)
    # for scalable k means pp algorithm, centroids = scalable_kpp_initial_centroids(data, k, l)[0]
    # data = scalable_kpp_initial_centroids(data, k, l)[1]
    stable = False
    j = []
    loopCount = 0
    while (not stable) & (loopCount < 1000):
        # find the Euclidean distance between a center and a data point
        data_ex = data[:, np.newaxis, :]
        euclidean_dist = (data_ex - centroids) ** 2
        distance = np.sum(euclidean_dist, axis = 2)
            
        # find to which cluster each data point belongs
        # create a matrix min_id, and let min_id[i,j] = 1 if the ith data point belongs to cluster j
        min_id = np.zeros(distance.shape)
        min_id[range(distance.shape[0]), np.argmin(distance, axis = 1)] = 1
            
        # calculate J
        j_v = np.sum(distance[min_id == True])
        j.append(j_v)
            
        # calculates the new centroids
        new_centroids = np.empty(centroids.shape)
        for c in range(0, k):
            if data[min_id[:, c] == True, :].shape[0] == 0:
                new_centroids[c] = centroids[c]
            else:
                new_centroids[c] = np.mean(data[min_id[:, c] == True, :], axis=0)
            
        # compare centroids to see if converge
        if compare_centroids(centroids, new_centroids):
            # If this is TRUE, then it has resulted in the same centroids
            stable = True
        else:
            centroids = new_centroids

        loopCount += 1
                
    return loopCount, j, centroids, min_id
```

###Application and comparison
####Simulate data to show how it works

```python
# Simulate data
data = np.random.randn(100000, 2)
k = 3
l = 4
# Show the results
# Basic k means
print ('Basic K means algorithm:')
kmeans_result = kmeans(k_initial_centroids(data, k), data, k)
print 'Using basic K means algorithm, after', kmeans_result[0], 'iterations, the algorithm stabilized.'
print 'Using basic K means algorithm, the centroids are', kmeans_result[2]
print 
# K means pp
print ('K means ++ algorithm:')
kpp_result = kmeans(kpp_initial_centroids(data, k), data, k)
print 'Using K means ++ algorithm, after', kpp_result[0], 'iterations, the algorithm stabilized.'
print 'Using K means ++ algorithm, the centroids are', kpp_result[2]
print 
# Scalable K means PP
print ('Scalable K means ++ algorithm:')
skpp_result = kmeans(scalable_kpp_initial_centroids(data, k, l)[0], scalable_kpp_initial_centroids(data, k, l)[1], k)
print 'Using Scalable K means ++ algorithm, after', skpp_result[0], 'iterations, the algorithm stabilized.'
print 'Using Scalable K means ++ algorithm, the centroids are', skpp_result[2]
```

####Comparison

```python
import timeit

def samplesize_initcost():
    """Compare the time to generate initial centers"""
    k = 10
    l = 10
    sampleSizes = [1000, 2000, 3000, 4000, 5000, 6000]
    km_time_init = []
    kmpp_time_init = []
    skmpp_time_init = []
    for n in sampleSizes:
        data = np.random.randn(n, 2)
        
        t1 = timeit.Timer(lambda: k_initial_centroids(data, k))
        km_time_init.append(t1.timeit(number = 10) * 100)
        
        t2 = timeit.Timer(lambda: kpp_initial_centroids(data, k))
        kmpp_time_init.append(t2.timeit(number = 10) * 100)

        t3 = timeit.Timer(lambda: scalable_kpp_initial_centroids(data, k, l))
        skmpp_time_init.append(t3.timeit(number = 10) * 100)

    # plot
    plt.figure(figsize=(10, 10))
    plt.plot(sampleSizes, km_time_init, '-o', lw=3, markersize=10)
    plt.plot(sampleSizes, kmpp_time_init, '-o', lw=3, markersize=10)
    plt.plot(sampleSizes, skmpp_time_init, '-o', lw=3, markersize=10)
    plt.legend(('KMeans', 'KMeans++', 'Scalable KMeans++'), prop={'size': 18}, loc=0)
    plt.xlabel('Sample Size', fontsize=18)
    plt.ylabel('Initial Cost (ms)', fontsize=18)
    ax = plt.gca()
    ax.xaxis.set_tick_params(labelsize=16)
    ax.xaxis.get_major_formatter().set_powerlimits((0, 0))
    ax.yaxis.set_tick_params(labelsize=16)
    plt.show()

samplesize_initcost()
```

```python
def samplesize_Clustercost():
    """Compare the time to cluster data sets of different sizes"""
    k = 10
    l = 10
    sampleSizes = [1000, 2000, 3000, 4000, 5000, 6000]
    km_perf = []
    kmpp_perf = []
    skmpp_perf = []
    for n in sampleSizes:
        data = np.random.randn(n, 2)

        t1 = timeit.Timer(lambda: kmeans(k_initial_centroids(data, k), data, k))
        km_perf.append(t1.timeit(number = 10) * 100)
        
        t2 = timeit.Timer(lambda: kmeans(kpp_initial_centroids(data, k), data, k))
        kmpp_perf.append(t2.timeit(number = 10) * 100)

        t3 = timeit.Timer(lambda: kmeans(scalable_kpp_initial_centroids(data, k, l)[0], scalable_kpp_initial_centroids(data, k, l)[1], k))
        skmpp_perf.append(t3.timeit(number = 10) * 100)

    # plot
    plt.figure(figsize=(10, 10))
    plt.plot(sampleSizes, km_perf, '-o', lw=3, markersize=10)
    plt.plot(sampleSizes, kmpp_perf, '-o', lw=3, markersize=10)
    plt.plot(sampleSizes, skmpp_perf, '-o', lw=3, markersize=10)
    plt.legend(('KMeans', 'KMeans++', 'Scalable KMeans++'), prop={'size': 18}, loc=0)
    plt.xlabel('Sample Size', fontsize=18)
    plt.ylabel('Cluster Cost (ms)', fontsize=18)
    ax = plt.gca()
    ax.xaxis.set_tick_params(labelsize=16)
    ax.xaxis.get_major_formatter().set_powerlimits((0, 0))
    ax.yaxis.set_tick_params(labelsize=16)
    plt.show()

samplesize_Clustercost()
```

```python
def samplesize_iteration():
    """Compare the number of iterations to cluster data sets of different sizes"""
    k = 10
    l = 10
    sampleSizes = [1000, 2000, 3000, 4000, 5000, 6000]
    km_iteration = []
    kmpp_iteration = []
    skmpp_iteration = []
    for n in sampleSizes:
        data = np.random.randn(n, 2)
        i1 = kmeans(k_initial_centroids(data, k), data, k)[0]
        i2 = kmeans(kpp_initial_centroids(data, k), data, k)[0]
        i3 = kmeans(scalable_kpp_initial_centroids(data, k, l)[0], scalable_kpp_initial_centroids(data, k, l)[1], k)[0]
        km_iteration.append(i1)  
        kmpp_iteration.append(i2)
        skmpp_iteration.append(i3)

    # plot
    plt.figure(figsize=(10, 10))
    plt.plot(sampleSizes, km_iteration, '-o', lw=3, markersize=10)
    plt.plot(sampleSizes, kmpp_iteration, '-o', lw=3, markersize=10)
    plt.plot(sampleSizes, skmpp_iteration, '-o', lw=3, markersize=10)
    plt.legend(('KMeans', 'KMeans++', 'Scalable KMeans++'), prop={'size': 18}, loc=0)
    plt.xlabel('Sample Size', fontsize=18)
    plt.ylabel('Iterations', fontsize=18)
    ax = plt.gca()
    ax.xaxis.set_tick_params(labelsize=16)
    ax.xaxis.get_major_formatter().set_powerlimits((0, 0))
    ax.yaxis.set_tick_params(labelsize=16)
    #plt.savefig('samples-initcost.png')
    #plt.close()
    plt.show()

samplesize_iteration()
```
